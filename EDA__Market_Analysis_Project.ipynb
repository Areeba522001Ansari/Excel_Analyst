```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_grp=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_grp,hue='Gender')
```

above graph represent that the category who purchased the product  were female

```python
# toal amount/ Age Group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the above graph most of the buyers are women between 26-35

```python
Sales_states=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
```

```python
sns.set(rc={'figure.figsize':(30,5)}) # width =35, height=5
```

```python
sns.barplot(data=Sales_states,x='State',y='Orders')
```

From the above graphs, this proves that Up has the higest order purchased by women

```python
# amount and states
Sales_states=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(30,5)}) # width =35, height=5
sns.barplot(data=Sales_states,x='State',y='Amount')
```

```python
# Marital status
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(2,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Sales_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)}) # width =35, height=5
sns.barplot(data=Sales_marital,x='Marital_Status',y='Amount',hue='Gender')
```

```python
# occupation
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# amount vs Occupation
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_occupation,x="Occupation",y="Amount")
```

Most of the buyers are From the IT Sector

```python
sns.set(rc={'figure.figsize':(28,5)})
sns.countplot(data=df,x='Product_Category')
for i in ax.containers:
  ax.bar_label(bars)
```

```python
sales_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(34,5)})
sns.barplot(data=sales_pc,x="Product_Category",y="Amount")
```

most sold product food , clothing, electronics

```python
sales_pro=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sns.set(rc={'figure.figsize':(34,5)})
sns.barplot(data=sales_pro,x="Product_ID",y="Orders")
```

Conclusion : Women age group 26-35 are the most buyers from UP Maharashtra, karnatka, working in IT , Aviation ,HealthCare companys buy the product From Food, clothing and Electronics category.
